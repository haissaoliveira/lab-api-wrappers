```python
artist_name = "Anitta"
results = sp.search(q=f'artist:{artist_name}', type='artist')
artist_id = results['artists']['items'][0]['id']
```

```python
top_tracks = sp.artist_top_tracks(artist_id)
top_tracks
```

```python
for idx, track in enumerate(top_tracks['tracks'][:10]):
    print(f"{idx + 1}. {track['name']} - {track['popularity']}")
```

```python
for idx, track in enumerate(top_tracks['tracks'][:10]):
    print(f"Track {idx + 1}:")
    print(f"Name: {track['name']}")
    print(f"Album: {track['album']['name']}")
    print(f"Release Date: {track['album']['release_date']}")
    print(f"Popularity: {track['popularity']}")
    print(f"Preview URL: {track['preview_url']}")
    print(f"Spotify URL: {track['external_urls']['spotify']}")
    print("-" * 40)
```
